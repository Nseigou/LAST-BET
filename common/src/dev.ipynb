```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import re
import time
from selenium.webdriver.chrome.options import Options


def scrape_kaisai_date_list(kaisai_date_list):
    options = Options()
    options.add_argument('--headless')
    race_id_list = []
    
    with webdriver.Chrome(options=options) as driver:
        for kaisai_date in tqdm(kaisai_date_list):
            url = f"https://race.netkeiba.com/top/race_list.html?kaisai_date={kaisai_date}"
            try:
                driver.get(url)
                time.sleep(1)
                li_list = driver.find_elements(By.CSS_SELECTOR, ".RaceList_DataItem")
                for li in li_list:
                    href = li.find_element(By.TAG_NAME, "a").get_attribute("href")
                    race_id = re.findall(r"race_id=(\d{12})", href)[0]
                    race_id_list.append(race_id)
            except :
                print(f"Error: {race_id}")
                print(traceback.format_exc())
                break
    return race_id_list
        
        
```

# 関数化

```python
from pathlib import Path
import pickle
from urllib.request import Request, urlopen
import time
from tqdm import tqdm  # tqdm をインポート

def scrape_race_html(race_id_list):
    file_path = Path("../data/race_id/race_id_list[2024-01-01 to 2025-01-31].pkl")
    with open(file_path, "rb") as f:
        race_id_list = pickle.load(f)

    html_dir = Path("../data/html/")
    html_dir.mkdir(parents=True, exist_ok=True)

    html_list = []
    for race_id in tqdm(race_id_list):
        url = f"https://db.netkeiba.com/race/{race_id}"
        html_file_path = html_dir / f"{race_id}.bin"
        req = Request(url, headers={"User-Agent": "Mozilla/5.0"})
        # ファイルが存在する場合はスキップ
        if html_file_path.exists():
            print(f"Skip: file exists ({html_file_path})")
        # ファイルが存在しない場合は取得
        else:
            time.sleep(1)
            html = urlopen(req).read()
            html_list.append(html)
            # ファイルに書き込み（書き込みモード "wb" を使用）
            with open(html_file_path, "wb") as f:
                f.write(html)
    return html_list
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import re

def create_race_results(html_race_file_list):
    dfs = {}
    error_race_ids = []
    
    for html_race_file in html_race_file_list:
        race_id = html_race_file.stem
        try:
            with open(html_race_file, "rb") as f:
                html = f.read()
                
                # BeautifulSoupを使用してレース結果テーブルを直接抽出
                soup = BeautifulSoup(html, 'html.parser')
                target_table = soup.find('table', class_='race_table_01')
                
                if not target_table:
                    print(f"No race result table found in race_id {race_id}")
                    error_race_ids.append(race_id)
                    continue
                
                # HTMLをUTF-8に変換して再度pd.read_htmlを試行
                html_str = str(target_table).encode('utf-8')
                try:
                    df = pd.read_html(html_str, encoding='utf-8')[0]
                    df.index = [race_id] * len(df)
                    dfs[race_id] = df
                except Exception as table_e:
                    print(f"Failed to parse table in race_id {race_id}")
                    print(f"Error: {str(table_e)}")
                    error_race_ids.append(race_id)
                    continue
                
        except Exception as e:
            error_race_ids.append(race_id)
            print(f"Error in race_id {race_id}")
            print(f"Error type: {type(e).__name__}")
            print(f"Error message: {str(e)}")
            continue
    
    if error_race_ids:
        print("\nエラーが発生したrace_idの数:", len(error_race_ids))
        print("\nエラーが発生したrace_id一覧:")
        for race_id in error_race_ids:
            print(race_id)
    
    if dfs:
        concat_df = pd.concat(dfs.values())
        concat_df.index.name = "race_id"
        return concat_df
    return None

# エラーが発生したファイルの詳細を確認する関数
def check_race_table(race_id, html_race_file_list):
    target_file = next((f for f in html_race_file_list if f.stem == race_id), None)
    if target_file:
        with open(target_file, "rb") as f:
            html = f.read()
            soup = BeautifulSoup(html, 'html.parser')
            target_table = soup.find('table', class_='race_table_01')
            
            if target_table:
                print("Found race result table:")
                print("Number of rows:", len(target_table.find_all('tr')))
                print("First few rows content:")
                for tr in target_table.find_all('tr')[:3]:
                    print([td.get_text(strip=True) for td in tr.find_all(['th', 'td'])])
            else:
                print("No race result table found")
```

```python
import pandas as pd
from pathlib import Path

# CSVファイルをタブ区切りとして読み込む
race_file_path = Path("../data/raw/csv/race_results/race_results.csv")
df = pd.read_csv(race_file_path, sep='\t')

# 単勝列を数値型に変換
df['単勝'] = pd.to_numeric(df['単勝'], errors='coerce')
df['馬番'] = pd.to_numeric(df['馬番'], errors='coerce')
df["人気"] = pd.to_numeric(df["人気"], errors="coerce")

# 条件に合う馬を抽出
target_horses = df[
    (df['性齢'].str.startswith('牝')) &  # 牝馬
    (df['馬番'] <= 4) &                  # 馬番4以内
    (df['人気'] == 1)                    # 1番人気
]

# 1着の数を計算
winners = target_horses[target_horses['着順'] == '1']
total_races = len(target_horses)
winning_races = len(winners)

# 勝率を計算
win_rate = (winning_races / total_races * 100) if total_races > 0 else 0

print(f"牝馬かつ馬番４以内かつ１番人気: {total_races}")
print(f"1着の数: {winning_races}")
print(f"勝率: {win_rate:.2f}%")

# 詳細情報の表示
print("\n勝利した馬の詳細:")
if winning_races > 0:
    columns_to_show = ['race_id', '馬番', '馬名', '性齢', '単勝', 'タイム']
    print(winners[columns_to_show])
```
